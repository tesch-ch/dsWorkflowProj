```python
# Get longest text in text column
longest_text = df2[df2['text'].str.len() == df2['text'].str.len().max()]['text'].iloc[0]
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained("Softechlb/articles_classification")
model = AutoModelForSequenceClassification.from_pretrained("Softechlb/articles_classification")

```

```python
# Tokenize input text
inputs = tokenizer(longest_text, padding=True, truncation=True, max_length=512, return_tensors="pt")

# Make prediction
outputs = model(inputs["input_ids"], attention_mask=inputs["attention_mask"])
predicted_label = torch.argmax(outputs.logits)

outputs.logits, model.config.id2label[predicted_label.item()]
```

```python
# Tokenize input text
inputs = tokenizer(longest_text, padding=True, truncation=True, max_length=512, return_tensors="pt")

# Make prediction
outputs = model(inputs["input_ids"], attention_mask=inputs["attention_mask"])
# outputs = model(**inputs)

# Get predicted class probabilities
predicted_probabilities = torch.softmax(outputs.logits, dim=1)

# Get class labels and corresponding probabilities
class_labels = model.config.id2label.values()
class_probabilities = predicted_probabilities[0].tolist()  # Assuming batch size is 1

# Create a dictionary. Class label keys, values are respective probabilities
class_probabilities_dict = dict(zip(class_labels, class_probabilities))

# Find the class label with the highest probability
predicted_class = max(class_probabilities_dict, key=class_probabilities_dict.get)

predicted_class, class_probabilities_dict
```

```python
class ProgressLog:
    """Prints the actual progress, each time after 1 % of the total is
    completed.
    """
    def __init__(self, total):
        self.total = total
        self.interval_milestone = 1
        self.current_count = 0

    def increase(self):
        self.current_count += 1
        progress = (self.current_count / self.total) * 100
        if progress >= self.interval_milestone:
            print(f"{self.current_count}/{self.total} -> {progress:.0f}% completed")
            self.interval_milestone += 1  # Increment the next progress milestone

# Example usage:
total_elements = 10000

# Create a progress bar
progress_bar = ProgressLog(total_elements)

# Simulate a loop
for i in range(total_elements):
    progress_bar.increase()
```

```python
from transformers import (AutoTokenizer, AutoModelForTokenClassification,
                          pipeline)
# from transformers import pipeline

tokenizer = AutoTokenizer.from_pretrained("dslim/bert-base-NER")
model = AutoModelForTokenClassification.from_pretrained("dslim/bert-base-NER")
```

```python
nlp = pipeline("ner", model=model, tokenizer=tokenizer, aggregation_strategy="simple")
example = "My name is Clara Berger and I live in Berkeley, California."
example = longest_text

ner_results = nlp(example)
ner_results
```
